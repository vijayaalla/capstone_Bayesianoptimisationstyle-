# Stage 2 Capstone: Black-Box Function Optimisation

This project simulates a Bayesian optimisation challenge across eight unknown functions. 
You do not know each function equation in advance. You start from limited observations and must select one new query point per function each round to maximise output.

The aim is not perfect global optimisation. The aim is strong, data-driven decision making with clear reflection on your strategy.

## What You Need To Do Each Round

1. Review current observations for each function.
2. Choose one new input point for each function.
3. Submit points in the required portal format.
4. Receive new outputs and append to your dataset.
5. Reflect on exploration vs exploitation and update your plan for the next round.

Allowed approaches include random search, grid search, Bayesian optimisation (GP + acquisition), manual reasoning, or custom surrogate models.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
```

```python
DATA_DIR = Path('../initial_data')

def load_all_functions(data_dir=DATA_DIR):
    """Load initial inputs/outputs for function_1 ... function_8."""
    all_data = {}
    for func_dir in sorted(data_dir.glob('function_*')):
        func_id = int(func_dir.name.split('_')[1])
        x_path = func_dir / 'initial_inputs.npy'
        y_path = func_dir / 'initial_outputs.npy'

        if not x_path.exists() or not y_path.exists():
            continue

        X = np.load(x_path)
        y = np.load(y_path)
        all_data[func_id] = {'X': X, 'y': y}

    return all_data

functions = load_all_functions()

for i in sorted(functions):
    X = functions[i]['X']
    y = functions[i]['y']
    print(f'function_{i}: X{X.shape}, y{y.shape}')
```

```python
def plot_function(func_id, X, y):
    dim = X.shape[1]
    print(f'Function {func_id} | dim={dim} | points={X.shape[0]}')

    if dim == 2:
        plt.figure(figsize=(6, 5))
        plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=70)
        plt.colorbar(label='Output')
        plt.title(f'Function {func_id}: 2D Input Space')
        plt.xlabel('x1')
        plt.ylabel('x2')
        plt.show()

    elif dim == 3:
        fig = plt.figure(figsize=(7, 6))
        ax = fig.add_subplot(111, projection='3d')
        p = ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis', s=70)
        fig.colorbar(p, label='Output')
        ax.set_title(f'Function {func_id}: 3D Input Space')
        ax.set_xlabel('x1')
        ax.set_ylabel('x2')
        ax.set_zlabel('x3')
        plt.show()

    else:
        # For 4D+ inputs, visualize a 2D PCA projection.
        pca = PCA(n_components=2)
        X_pca = pca.fit_transform(X)

        plt.figure(figsize=(6, 5))
        plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', s=70)
        plt.colorbar(label='Output')
        plt.title(f'Function {func_id}: PCA Projection ({dim}D -> 2D)')
        plt.xlabel('PC1')
        plt.ylabel('PC2')
        plt.show()
```

```python
for i in sorted(functions):
    plot_function(i, functions[i]['X'], functions[i]['y'])
```
